### Veriyi Değiştirme


```
# df["A"].str.lower()
# policeDf["violation"].str.upper()
# policeDf.isna().sum()
```

```
# df.fillna()
# 1
# yeniSutun =  policeDf["driver_age"].fillna(policeDf["driver_age"].mean())
policeDf["driver_age"].fillna(policeDf["driver_age"].mean(),inplace=True)
policeDf.isna().sum()
policeDf["driver_age"] =  policeDf["driver_age"].astype(int)
policeDf.info()
```

```
# df.dropna()
policeDf.isna().sum()
# policeDf["driver_age_raw"].dropna()
dfNetflix_Na_temiz = dfNetflix.copy()
len(dfNetflix)
dfNetflix_Na_temiz.dropna(inplace=True)
dfNetflix_Na_temiz.info()
```

```
# Yeni bir seri üretme DataFrame üzerine ekleme
seri1 = pd.Series(np.random.randint(20,size=(3774,)))
dfNetflix_Na_temiz["benim"] = seri1
dfNetflix_Na_temiz.isna().sum()
```

```
# Hepsini aynı değere ayarlama
dfNetflix_Na_temiz["benim"] = 5
dfNetflix_Na_temiz
```

```
# reset_index()
dfNetflix_Na_temiz.reset_index(drop=True,inplace=True)
seri1 = pd.Series(np.random.randint(20,size=(3774,)))
dfNetflix_Na_temiz["benim"] = seri1
dfNetflix_Na_temiz
```

```
# Örnek Çıkarma - sample
dfNetflix_Na_temiz_ornek = dfNetflix_Na_temiz.sample(frac=0.2)
dfNetflix_Na_temiz_ornek 
```

```
dfNetflix_Na_temiz =  dfNetflix_Na_temiz.drop("benim",axis=1)
dfNetflix_Na_temiz
```

```
# .apply()
dfNetflix_Na_temiz["Fonksiyon"] = dfNetflix_Na_temiz["release_year"].apply(lambda x: x-1900)
dfNetflix_Na_temiz 
```
